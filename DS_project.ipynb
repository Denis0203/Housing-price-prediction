```python
#функция построения гистограммы
def draw_hist(data,column):
    fig = px.histogram(data,nbins = 100)
    fig.update_layout(
    title= column + ' hist',
    xaxis_title=column,
    yaxis_title="Count")
    fig.show()

#функция очистки от выбросов с помощью межквантильного диапазона     
def iqr_check(data,column):
    IQR = data[column].describe()['75%'] - data[column].describe()['25%']
    min_limit = data[column].describe()['25%'] - IQR
    max_limit = data[column].describe()['75%'] + IQR
    data = data[(data[column]>min_limit)&(data[column]<max_limit)]
    return data

def fill_nan(data):
    LifeSquare_coef = (data[data['LifeSquare'].isna()==False]['LifeSquare']/data[data['LifeSquare'].isna()==False]['Square']).mean()
    data.loc[data['LifeSquare'].isna()==True,'LifeSquare']=data[data['LifeSquare'].isna()==True]['Square']*LifeSquare_coef
    data['Healthcare_1'] = data['Healthcare_1'].fillna(data['Healthcare_1'].median())
    cat_ddataict = {'A':1,'B':2}
    data['Ecology_2'].replace(cat_dict, inplace=True)
    data['Ecology_3'].replace(cat_dict, inplace=True)
    data['Shops_2'].replace(cat_dict, inplace=True)
    return data
```

2 Обучение модели

Разделим набор на параметры и целевое значение

```python
x = train_data.drop('Price', axis=1)
y = train_data['Price'].tolist()
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size=0.2, random_state=42,shuffle = True)
```

```python
model = RandomForestRegressor( random_state=21, criterion='squared_error')
model.fit(x_train, y_train)
```

```python
y_train_preds = model.predict(x_train)
y_test_preds = model.predict(x_valid)

r2 = r2_score(y_train, y_train_preds)

print("Train "+f"R²: {r2}")

r2 = r2_score(y_valid, y_test_preds)

print("Valid "+f"R²: {r2}")
```

```python
cv_score = cross_val_score(model,x_train,y_train, scoring='r2', cv=KFold(n_splits=3,shuffle = True,random_state=21))
cv_score
```
